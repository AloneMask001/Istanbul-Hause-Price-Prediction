```python
from sklearn.preprocessing import LabelEncoder
import re
```

```python
#burası Kat sayısını numeric değere çeviren yer.
def kat_cevirici (kat_girdisi):
    
    ozel_kat_degerler = {
    'Bahçe Katı': 0, 'Düz Giriş': 0, 'Yüksek Giriş': 0,
    'Tam Bodrum': -1, 'Kot 1 (-1). Kat': -1,
    'Kot 2 (-2). Kat': -2, 'Kot 3 (-3). Kat': -3,
    '30-40. Kat': 35, '10-20. Kat': 15,
    'Villa Tipi': 2, 'Müstakil': 2,
    'Çatı Katı': 'CATI', 'Çatı Dubleks': 'CATI'
}
    kat_sayısı_hesaplama = re.search(r'\d+', str(kat_girdisi))
    
    if kat_girdisi in ozel_kat_degerler:
        return ozel_kat_degerler.get(kat_girdisi)
    
    elif kat_sayısı_hesaplama:
        return int(kat_sayısı_hesaplama.group())
        
    
    
veri['Bulunduğu_Kat'] = veri['Bulunduğu_Kat'].apply(kat_cevirici)

max_kat = veri[veri['Bulunduğu_Kat'] != 'CATI']['Bulunduğu_Kat'].max()

veri.loc[veri['Bulunduğu_Kat'] == 'CATI', 'Bulunduğu_Kat'] = max_kat
 
```

```python
#Bina sayısı numeric yapma kısmı.

def bina_kat_numeric(bina_kat):
    ozel_kat_değerler_bina = {'5-10' : 7, '21 Ve Üzeri' : 25, '16-20' : 18, 
                     '0 (Yeni)': 0, '11-15': 13
                    }

    regex_bina_kat_cevirici = re.search(r'\d+', str(bina_kat))

    if bina_kat in ozel_kat_değerler_bina:
        return ozel_kat_değerler_bina.get(bina_kat)
    elif regex_bina_kat_cevirici:
        return int(regex_bina_kat_cevirici.group())

veri['Binanın_Yaşı'] = veri['Binanın_Yaşı'].apply(bina_kat_numeric)
```

```python
x = veri[['Binanın_Yaşı', 'Binanın_Kat_Sayısı', 'Kullanım_Durumu',
       'Yatırıma_Uygunluk', 'Banyo_Sayısı', 'Net_Metrekare', 'Oda_Sayısı',
       'Bulunduğu_Kat', 'Isıtma_Tipi', 'Krediye_Uygunluk', 'Site_İçerisinde',
        'yaka', 'Yaşam_endeksi', 'Nüfus', 'İlçe_Mahalle_target','kat_oran']]
y = np.log1p(veri['Fiyatı'])

# Fiyat dağılımını düzelt, uç değerleri sıkıştır-----np.log1p(veri['Fiyatı'])----*
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
```

```python
#Random forrest -> reduces overfitting. also bias
#We will descide what to put into the parameter

from sklearn.ensemble import RandomForestRegressor
```

```python
rfregressor = RandomForestRegressor(
    
    n_estimators=300,
    random_state=42,
    oob_score=True,
    max_depth = 25,
    min_samples_leaf = 2,
    max_features="sqrt"
)

rfregressor.fit(x_train, y_train)

# Out of Bag Error is a method used to estimate the prediction error of Random Forests.
```

```python
print("Out-of-Bag Score:", rfregressor.oob_score_)

#Bu log'lu fiyat tahmin zamanı bunu tekrardan normale çevirmemiz lazım!!!!
```

```python
print("Out-of-Bag Test Score:", rfregressor.score(x_test, y_test))
```
